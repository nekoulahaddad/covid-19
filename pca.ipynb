```python
## import libraries
import pandas as pd
import numpy as np
from feature_selector import FeatureSelector
import random as rd
from numpy import isnan
from sklearn.decomposition import PCA
from sklearn import preprocessing
import matplotlib.pyplot as plt # NOTE: This was tested with matplotlib v. 2.1.0
import sys
import openpyxl





 
```

```python
#reading data
data = pd.read_excel('itmo_catboost.xlsx', engine='openpyxl')
data = data.drop(['летальные'], axis=1)
#data = data.iloc[:98, :] #first group
#data = data.iloc[98:,:].reset_index(drop=True) #second group
number_of_rows = data.shape[0]
number_of_columns = data.shape[1]
#pd.set_option('display.max_columns',number_of_columns)
pd.set_option('display.max_rows',number_of_rows)
a = [x for x in range(number_of_rows)]
data.head(10)
```

```python
fs = FeatureSelector(data = data,labels=None)
fs.identify_missing(missing_threshold=0.6)
missing_features = fs.ops['missing']
missing_features[:10]
fs.plot_missing()
```

```python
fs.missing_stats.head(10)
```

```python
data = fs.remove(methods = ['missing'])
```

```python
number_of_columns = len(data.columns)
#features = ['feature' + str(i) for i in range(1,number_of_columns+1)]
features = list(data.columns) 
data.head(7)
```

```python
cols_fillna = list(data.columns) 
#first = cols_fillna[14]
#print(type(data[first]))
#ameen = data[first]
#ameen = ameen.values
#print(type(ameen))
#print(ameen)
#data.mean()  
#print(means_value)
#print(means_value[1])
#data[first].fillna(means_value[i].mean(),inplace=True)

means_value = data.mean()
for col in cols_fillna:
    data[col].fillna(means_value[data.columns.get_loc(col)].mean(),inplace=True) 

```

```python
X = data
data.head(5)
```

```python
scaled_data = preprocessing.scale(data)
pca = PCA(.95) # create a PCA object
pca.fit(scaled_data) # do the math
pca_data = pca.transform(scaled_data) # get PCA coordinates for scaled_data
```

```python
#PCA plot
per_var = np.round(pca.explained_variance_ratio_* 100, decimals=1)
labels = ['PC' + str(x) for x in range(1, len(per_var)+1)]
 
plt.bar(x=range(1,len(per_var)+1), height=per_var, tick_label=labels)
plt.ylabel('Percentage of Explained Variance')
plt.xlabel('Principal Component')
plt.title('Scree Plot')
plt.show()
```

```python
#plot using PC1 and PC2
pca_df = pd.DataFrame(pca_data, index=a, columns=labels)
 
plt.scatter(pca_df.PC1, pca_df.PC2)
plt.title('My PCA Graph')
plt.xlabel('PC1 - {0}%'.format(per_var[0]))
plt.ylabel('PC2 - {0}%'.format(per_var[1]))
 
for sample in pca_df.index:
    plt.annotate(sample, (pca_df.PC1.loc[sample], pca_df.PC2.loc[sample]))
 
plt.show()
```

```python
final_pca = pd.DataFrame(pca.components_.T, index=features)
PCAs_values = pd.Series(per_var) 
final_pca
```

```python
multiplicationResults = final_pca.mul(PCAs_values);
multiplicationResults
```

```python
mean_final = multiplicationResults.abs().mean(axis = 1).sort_values(ascending=False) 
```

```python
df1 = pd.DataFrame(data=mean_final[0:20].index, columns=['Best features for second group'])
df2 = pd.DataFrame(data=mean_final[0:20].values, columns=['Best features values'])
best_features_with_PCA = pd.merge(df1, df2, left_index=True, right_index=True)
```

```python
best_features_with_PCA
```
