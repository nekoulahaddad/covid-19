```python
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
from feature_selector import FeatureSelector
from sklearn.model_selection import train_test_split

import catboost
from catboost import CatBoostClassifier 
```

```python
data = pd.read_excel('itmo_catboost.xlsx', engine='openpyxl')
#data = data.iloc[:98, :] #first group
#data = data.iloc[98:,:].reset_index(drop=True) #second group
number_of_rows = data.shape[0]
number_of_columns = data.shape[1]
#pd.set_option('display.max_columns',number_of_columns)
pd.set_option('display.max_rows',number_of_rows)
a = [x for x in range(number_of_rows)]
data.head(10)
```

```python
fs = FeatureSelector(data = data,labels=None)
fs.identify_missing(missing_threshold=0.6)
missing_features = fs.ops['missing']
missing_features[:10]
fs.plot_missing()
```

```python
fs.missing_stats.head(10)
```

```python
data = fs.remove(methods = ['missing'])
data.head(7)
```

```python
cols_fillna = list(data.columns) 
means_value = data.mean()
for col in cols_fillna:
    data[col].fillna(means_value[data.columns.get_loc(col)].mean(),inplace=True) 

```

```python
data.head(5)
```

```python
cor = data.corr()
cor
```

```python
import seaborn as sns
plt.figure(figsize=(10,6))
sns.heatmap(cor,annot=True) 
```

```python
features = [feat for feat in list(data) 
            if feat != 'летальные']

categorical_features = np.where(data[features].dtypes != float)[0]
 

X_train, X_test, y_train, y_test = train_test_split(data[features], 
                                                    data[['летальные']], 
                                                    test_size=0.3, 
                                                     random_state=1)
 
params = {'iterations':5000,
        'learning_rate':0.01,
        'cat_features':categorical_features,
        'depth':3,
        'eval_metric':'AUC',
        'verbose':200,
        'od_type':"Iter", 
        'od_wait':500, 
        'random_seed': 1
          }

cat_model = CatBoostClassifier(**params)
cat_model.fit(X_train, y_train,   
          eval_set=(X_test, y_test), 
          use_best_model=True, 
          plot=True  
         );
```

```python
#plot best features (first way)
import shap  # package used to calculate Shap values

from catboost import CatBoostClassifier, Pool
shap_values = cat_model.get_feature_importance(Pool(X_test, label=y_test,cat_features=categorical_features) ,
                                               type="ShapValues")
 
expected_value = shap_values[0,-1]
shap_values = shap_values[:,:-1]

shap.initjs()
shap.force_plot(expected_value, shap_values[0,:], X_test.iloc[0,:])
```

```python
#plot best features (second way)
shap.summary_plot(shap_values, X_test)
```
